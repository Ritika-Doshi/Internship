```python
##.Scrape the details most watched tv series of all time from imdb.com. 
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page= requests.get("https://www.imdb.com/list/ls512407256/")
soup = BeautifulSoup(page.text, 'html.parser')

try:
    series_list = soup.find_all('div', class_='lister-item-content')

    names = []
    year_spans = []
    genres = []
    run_times = []
    ratings = []
    votes = []

    for series in series_list:
        name_elem = series.find('h3', class_='lister-item-header').find('a')
        name = name_elem.text.strip() if name_elem else None
        names.append(name)

        year_span_elem = series.find('span', class_='lister-item-year')
        year_span = year_span_elem.text.strip('()') if year_span_elem else None
        year_spans.append(year_span)
        
        genre_elem = series.find('span', class_='genre')
        genre = genre_elem.text.strip() if genre_elem else None
        genres.append(genre)
       
        run_time_elem = series.find('span', class_='runtime')
        run_time = run_time_elem.text.strip() if run_time_elem else None
        run_times.append(run_time)
        
        rating_elem = series.find('strong')
        rating = rating_elem.text.strip() if rating_elem else None
        ratings.append(ratings)
        
        vote_elem = series.find('span', attrs={'name': 'nv'})
        vote = vote_elem.text.strip() if vote_elem else None
        votes.append(vote)
        
    df = pd.DataFrame({"Name": names,"Year Span": year_spans,"Genre": genres,"Run Time": run_times,"Ratings": ratings,"Votes": votes})
    print(df)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
!pip install selenium
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome() 
driver.get("https://archive.ics.uci.edu")

datasets=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
datasets.click()

driver.implicitly_wait(10)

urls=[]    
dataset_names = []
data_types_ = []
tasks = []
attribute_types = []
no_of_instances = []
no_of_attributes = []

url_tags=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in url_tags:
    url=i.get_attribute('href')
    urls.append(url)

for url in urls:
    driver.get(url)
    dataset_name_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//div[@class="flex items-center gap-4"]/h1'))
    )
    dataset_names.append(dataset_name_element.text)

    
dataset_names
```

```python
for url in urls:
    driver.get(url)
    data_type=driver.find_element(By.XPATH,'//p[@class="text-md"]')
    data_types_.append(data_type.text)
    
data_types
```

```python
for url in urls:
    driver.get(url)
    task=driver.find_element(By.XPATH,'//div[@class="col-span-4"][3]/p')
    tasks.append(task.text)
    
    attribute=driver.find_element(By.XPATH,'//div[@class="col-span-4"][4]/p')
    attribute_types.append(attribute.text)

    instances=driver.find_element(By.XPATH,'//div[@class="col-span-4"][5]/p')
    no_of_instances.append(instances.text)

    attributes=driver.find_element(By.XPATH,'//div[@class="col-span-4"][6]/p')
    no_of_attributes.append(attributes.text)
    
```

```python
print(len(tasks),len(attribute_types),len(no_of_instances),len(no_of_attributes))
```

```python
print(f"Dataset: {dataset_names}\nADatatype: {data_types}\nTask: {tasks}\nAttribute_type: {attribute_types}\nNo of instances: {no_of_instances}\nNo of attributes: {no_of_attributes}\n---")
```
